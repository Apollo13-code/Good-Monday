```python
import random

while True:
    message = input("Would you like to roll the dice? Please enter Yes/No")
    if message.lower() == "no":
        break
    elif message.lower == "yes":
        dice_number = random.randint(1,6)
        print(f"The number you've rolled is {dice_number}")
    else:
        print("The answer must be yes or no")
    
```
